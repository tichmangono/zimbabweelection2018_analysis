## PART6 - Introducing "Parly Ratio" - how well or badly parliamentary candidates performed relative to the presidential candidate in the same party for one constituency
--------

#### Definition:
**Parly Ratio** = Parliamentary votes as a percentage of presidential vote for a particular party in a particular constituency, province or nation. 
- I just made up the above term to make it easier to talk about this number 
- For example, in one constituency:
    - MDC Alliance may have 9 parliamentary votes and 10 presidential votes
    - Zanu PF may have 19 parliamentary votes and 20 presidential votes
    - Parly Ratio is 9/10= 90% for MDC and 19/20 = 95% for Zanu PF in that particular constituency

#### Parly Ratio interpretation
 - if **LOW** generally means that the parlimentary candidate for that party got **LESS** votes than the presidential candidate for the same party in that constituency
 - if **HIGH** generally means that the parlimentary candidate for that party got **MORE** votes than the presidential candidate for the same party in that constituency

**Note:** This is just a hypothesis at the moment, just allow some exploration of the differences in parliamentary and presidential votes by political party. It is possible that there are alternative explanations for the follwoing differences and phenomena (other than election fraud or rigging)
- **Assumption 1: Size of parly ratio should be close to 100%** That, in a harmonized election we would expect these numbers to be close to each other i.e. the number of votes that an MDC presidential candidate would get in a constituency for is close to the number of votes an MDC parliamentary candidate would get in the same constituency give or take some small %. So this number would maybe range from 90% to 110% of presidential, just to take the extreme cases, where we give a margin of about 10%
- **Assumption 2: Location and distribution of abnomarly variant parly ratios i.e. less than 90% or more than 110% of presidential should not be pervasive, i.e. it will be concentrated in certain area where a party is particularly strong, or particularly weak for example** - Even if there are big margins, these margins would not be expected to be high in every constituency, we should see a random distribution. However, it seems this was not the case. 
 
That said let's look at the distribution of this measure across political parties and province.

#### Analyzing the "parly ratio" data shows that:
- Zanu PF had national average parly ratio that was almost perfect at **99.6%.**
- At the same time, MDC's average national parly ratio was very low at **76%**, (with a 24% deviation from the mean 100%)
- Out of 210 constituencies, Zanu PF consistently had a higher parly ratio,  **at 190 out of 210 constituencies.** That is **over 90% ** of all the constituencies
- If we got polling station data for the parliamentary election, we could do further anlaysis
    - It seems unlikely that at almost every poling station, an average of 24% of MDC supports who would vote forMDC Alliance president would then turn around and either vote for Zanu PF or other candidates.
    - A possible explanation is that MDC internal politics (e.g. imposed candidates and the Khupe factor) could have pulled supporters away from the MDC or split the vote. But since Khupe and all others only got less than 5% of the presidential vote, it hard to imagine how the other 20% would then be distributed
    - Additionally, we expected Zanu PF to also suffer from a split factor at some polling stations e.g. the G40 factor, would newtralize the Khupe factor at the very least
    - The fact that we are seeing marked differences and than Zanu PF has higher parly ratio than MDC Allicance at almost all constituencies sems worthy of investigating
    
#### Zanu PF parly ratios are skewed to higher levels while MDC Alliance parly ratios are on the lower side. The distribution of MDC parly ratios also doesn't follow the expected normal distribution. This may just reflect that Chamisa was significantly more popular than his MP candidates, so needs further analysis

#### Notice that:
- Zanu PF parly ratio distribution is roughly normally distributed around the mean (and high peak) at 1 or 100%
- BUT the MDC parly ratios are significantly skewed to the left and generally smaller. This suggests possible anomalies or very extraordinary explanations. Both the size of the parly ratios and their distributions violate the assumptions 1 and 2 made above

#### Zanu PF has high parly ratio in 37 (18%) of all constituencies (and MDC has high parly ratio in ony 1 constituency)

## Introducing "Parly Ratio" measure/calculation
- Parly Ratio = Parliamentary votes as a percentage of presidential vote for a particular party in a particular constituency, province or nation. 
- I just made up the above term to make it easier to talk about this number 
- For example, on constituency:
    - MDC Alliance may have 9 parliamentary votes and 10 presidential votes
    - Zanu PF may have 19 parliamentary votes and 20 presidential votes
    - Parly Ratio is 9/10= 90% for MDC and 19/20 = 95% for Zanu PF in that particular constituency
    
Note: This is just a hypothesis at the moment, just allow some exploration of the differences in parliamentary and presidential votes by political party. It is possible that there are alternative explanations for the follwoing differences and phenomena (other than election fraud or rigging)
- **Assumption 1: Size of parly ratio should be close to 100%** That, in a harmonized election we would expect these numbers to be close to each other i.e. the number of votes that an MDC presidential candidate would get in a constituency for is close to the number of votes an MDC parliamentary candidate would get in the same constituency give or take some small %. So this number would maybe range from 90% to 110% of presidential, just to take the extreme cases, where we give a margin of about 10%
- **Assumption 2: Location and distribution of abnomarly variant parly ratios i.e. less than 90% or more than 110% of presidential should not be pervasive, i.e. it will be concentrated in certain area where a party is particularly strong, or particularly weak for example** - Even if there are big margins, these margins would not be expected to be high in every constituency, we should see a random distribution. However, it seems this was not the case. 

#### Parly Ratio interpretation
 - if LOW generally means that the parlimentary candidate for that party got LESS votes than the presidential candidate for the same party in that constituency
 - if HIGH generally means that the parlimentary candidate for that party got MORE votes than the presidential candidate for the same party in that constituency
 
That said let's look at the distribution of this measure across political parties and province.


#### Analyzing the "parly ratio" data shows that:
- Zanu PF had national average parly ratio that was almost perfect at 99.6%. This is within the range
- At the same time, MDC's average national parly ratio was very low at 76%, (with a 24% deviation)
- Out of 210 constituencies, Zanu PF consistently had a higher parly ratio,  at 190 out of 210 constituencies. That is over 90% of all the constituencies
- If we got polling station data for the parliamentary election, we could do further anlaysis
    - It seems unlikely that at almost every poling station, an average of 24% of MDC supports who would vote forMDC Alliance president would then turn around and either vote for Zanu PF or other candidates.
    - A possible explanation is that MDC internal politics (e.g. imposed candidates and the Khupe factor) could have pulled supporters away from the MDC or split the vote. But since Khupe and all others only got less than 5% of the presidential vote, it hard to imagine how the other 20% would then be distributed
    - Additionally, we expected Zanu PF to also suffer from a split factor at some polling stations e.g. the G40 factor, would newtralize the Khupe factor at the very least
